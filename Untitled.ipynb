```python
from pdf2image import convert_from_path, pdfinfo_from_path
import pytesseract

def ocr_scrape_pdf_in_batches(file_path, batch_size=10):
    # 1. Get the total number of pages in the PDF without loading the images
    info = pdfinfo_from_path(file_path)
    total_pages = info["Pages"]
    print(f"Total pages found: {total_pages}")
    
    ocr_text = ""
    
    # 2. Loop through the document in chunks (e.g., 10 pages at a time)
    for start_page in range(1, total_pages + 1, batch_size):
        end_page = min(start_page + batch_size - 1, total_pages)
        print(f"Processing pages {start_page} to {end_page}...")
        
        # 3. Convert ONLY the current batch of pages to images
        pages = convert_from_path(file_path, dpi=300, first_page=start_page, last_page=end_page)
        
        # 4. Run OCR on this specific batch
        for i, page in enumerate(pages):
            text = pytesseract.image_to_string(page)
            ocr_text += f"--- PAGE {start_page + i} --- \n {text} \n"
            
        # At the end of this loop, Python will automatically free up the 
        # memory from the 'pages' variable before grabbing the next batch!

    return ocr_text

# Run the updated function
pdf_path = "Kimball Grammar.pdf"
ocr_grammar_text = ocr_scrape_pdf_in_batches(pdf_path, batch_size=10)
print("Finished!")
```
